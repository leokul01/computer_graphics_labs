```python
import matplotlib.pyplot as plt
from skimage.io import imread, imshow, imsave
from skimage.color import gray2rgb
from skimage.transform import rotate
from skimage import img_as_ubyte
import numpy as np
```

```python
data = imread('data.png')
data = gray2rgb(data)
data = img_as_ubyte(data)
```

```python
data_red = np.copy(data)

r1, g1, b1 = 200, 200, 200 # Original threshold for value
r2, g2, b2 = 254, 0, 0 # Value that we want to replace it with

red_c, green_c, blue_c = data_red[:,:,0], data_red[:,:,1], data_red[:,:,2]
mask = (red_c <= r1) & (green_c <= g1) & (blue_c <= b1)

data_red[:,:,:][mask] = [r2, g2, b2]

data_red = img_as_ubyte(data_red)
```

```python
imshow(data_red)
imsave('data_red.png', data_red)
```

```python
data_green = np.copy(data)

r1, g1, b1 = 200, 200, 200 # Original threshold for value
r2, g2, b2 = 0, 254, 0 # Value that we want to replace it with

red_c, green_c, blue_c = data_green[:,:,0], data_green[:,:,1], data_green[:,:,2]
mask = (red_c <= r1) & (green_c <= g1) & (blue_c <= b1)

data_green[:,:,:][mask] = [r2, g2, b2]

data_green = rotate(data_green, angle=90, preserve_range=True, mode='edge')

data_green = img_as_ubyte(data_green/255)
```

```python
imshow(data_green)
imsave('data_green.png', data_green)
```

```python
data_mix = (data_red + data_green) % 255

imsave('data_mix.png', data_mix)
imshow(data_mix)
```
